```python
import cv2
import sys
import glob
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def detect_face(path):
    imagePath = path
    cascPath = "haarcascade_frontalface_default.xml"

    # Create the haar cascade
    faceCascade = cv2.CascadeClassifier(cascPath)

    # Read the image
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.5,
        minNeighbors=5,
        minSize=(30, 30),
    )

    print("Found {0} faces!".format(len(faces)))

    # Draw a rectangle around the faces
    counter = 1
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        crop_img = image[y: y + h, x: x + w]
        newPath = os.path.join(os.path.join(os.getcwd(), "Cropped"), path)
        s = newPath.split('.')[0]
        #print s
        newPath = "{}-{}.jpg".format(s, counter)
        print newPath
        cv2.imwrite(newPath, crop_img)
        counter +=1
```

```python
ig_folderpath = 'Data/Images/Instagram/'
folders = os.listdir(ig_folderpath)
for folder_ in folders: 
    oldfolder = os.path.join(ig_folderpath, folder_)
    curfolder = "Cropped/"+oldfolder
    print curfolder
    if not os.path.exists(curfolder):
        os.makedirs(curfolder)
    oldfolder = oldfolder + '/*.jpg'
    print oldfolder
    _files = glob.glob(oldfolder)
    #print _files
    for f in _files:
        detect_face(f)
```
